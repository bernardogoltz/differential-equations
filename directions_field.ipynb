# Differencial Equations

```python
# importing the dependencies
import numpy as np 
import sympy as sp
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

The graphical representation of the solution of a Differential Equation is a Direction Field.

```python
# the vector at a point represents the magnitude of the field vector at that point.

def create_dfield(xsize , ysize, nticks):
    # arange(start,stop,step)
    x = np.arange(-1,xsize,nticks)
    y = np.arange(-0.5*ysize,ysize,nticks)

    X,Y = np.meshgrid(x,y)

    dy = X + np.sin(Y)
    dx = np.ones(dy.shape)

    return X,Y,dx,dy

x,y,u,v = create_dfield(4,2,np.pi*0.05)

plt.figure(figsize=(8,8))
plt.title("Direction Field a given differencial equation")
plt.quiver(x,y,u,v,
           color = 'purple',
           )
```

```python
# the vector at a point represents the magnitude of the field vector at that point.

def create_dfield(xsize , ysize, nticks):
    # arange(start,stop,step)
    x = np.arange(-xsize,xsize,nticks)
    y = np.arange(-0.5*ysize,ysize,nticks)

    X,Y = np.meshgrid(x,y)

    dy = X**3-Y
    dx = np.ones(dy.shape)

    return X,Y,dx,dy

x,y,u,v = create_dfield(3,5,0.5)

plt.figure(figsize=(8,8))
plt.title("Direction Field a given differencial equation")
plt.quiver(x,y,u,v,color = 'red')
```
